**Image To Pencil Sketch With Python**

**Import the library**

```python
import cv2
```

```python
img =cv2.imread('/content/banner-2.jpg');
```

```python
from google.colab.patches import cv2_imshow
```

***Read in the image***

```python
cv2_imshow(img);
```

***convert image to grey scale***

```python
gray_image=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

***invert the grey scale image***

```python
inverted_gray_image=255-gray_image
cv2_imshow(inverted_gray_image)
```

*** Blur the image by gaussian function***

```python
blurred_img = cv2.GaussianBlur(inverted_gray_image, (21,21), 0)
cv2_imshow(blurred_img)
```

***Invert the blurred image***

```python
inverted_blurred_image = 255-blurred_img
cv2_imshow(inverted_blurred_image)
```

***Create the pencil sketch image***

```python
pencil_sketch_image = cv2.divide(gray_image,inverted_blurred_image,scale=256.0)
cv2_imshow(pencil_sketch_image)
```
